```python
import torch
import numpy as np
from torch.utils.data import TensorDataset, DataLoader, ConcatDataset
```

```python
steps = 10
stride = 1
for i in range(1, 301):
    print(i)
    raw_np = np.load('../data/npy_4ch/sequence_{}.npy'.format(i))
    raw_tensor = torch.Tensor(raw_np)
    f = {}
    f[steps-1] = raw_tensor[steps-1:]
    for ii in range(steps-1):
        f[ii] = raw_tensor[ii*stride:-(steps-ii-1)*stride]
        # print(ii*stride,':',-(steps-ii-1)*stride)
        # print(len(f[ii]))
    tmp_dataset = TensorDataset(f[0],f[1],f[2],f[3],f[4],f[5],f[6],f[7],f[8],f[9])
    if i == 1:
        dataset = tmp_dataset
    else:
        dataset = ConcatDataset([dataset, tmp_dataset])
```

```python
torch.save(dataset, "../data/dataset_10steps_4ch.pt")
```

```python
import matplotlib.pyplot as plt

d1 = DataLoader(dataset, batch_size = 3, shuffle=True)

for idx, (x, y, z, _, _, _, _, _, _, _) in enumerate(d1):
    plt.imshow(x[1].permute(2,0,1)[:-1].permute(1,2,0).numpy()+0.5)
    plt.show()
    plt.imshow(y[1].numpy())
    plt.show()
    plt.imshow(z[1].numpy())
    plt.show()
    print(y[1])
    break
```
