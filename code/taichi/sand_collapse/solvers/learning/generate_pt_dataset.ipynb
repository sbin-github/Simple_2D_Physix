```python
import torch
import numpy as np
from torch.utils.data import TensorDataset, DataLoader, ConcatDataset
```

```python
for i in range(1, 301):
    raw_np = np.load('../data/sequence_{}.npy'.format(i))
    raw_tensor = torch.Tensor(raw_np)
    # print(x_np.shape)
    x_tensor = raw_tensor[:-1]
    y_tensor = raw_tensor[1:]
    tmp_dataset = TensorDataset(x_tensor, y_tensor)
    if i == 1:
        dataset = tmp_dataset
    else:
        dataset = ConcatDataset([dataset, tmp_dataset])
```

```python
torch.save(dataset, "../data/dataset.pt")
```

```python
import matplotlib.pyplot as plt

d1 = DataLoader(dataset, batch_size = 3, shuffle=True)

for idx, (x, y) in enumerate(d1):
    plt.imshow(x[1].numpy())
    plt.show()
    plt.imshow(y[1].numpy())
    break
```
