```python
import torch
import numpy as np
from torch.utils.data import TensorDataset, DataLoader, ConcatDataset
```

```python
steps = 10
stride = 3
for i in range(1, 301):
    raw_np = np.load('../data/sequence_{}.npy'.format(i))
    raw_tensor = torch.Tensor(raw_np)
    f = {}
    # f[steps-1] = raw_tensor[steps-1:]
    for ii in range(steps-1):
        f[ii] = raw_tensor[ii*stride:-(steps-ii)*stride]
        # print(len(f[ii]))
    tmp_dataset = TensorDataset(f[0],f[1],f[2],f[3],f[4],f[5])
    if i == 1:
        dataset = tmp_dataset
    else:
        dataset = ConcatDataset([dataset, tmp_dataset])
```

```python
torch.save(dataset, "../data/dataset_10steps.pt")
```

```python
import matplotlib.pyplot as plt

d1 = DataLoader(dataset, batch_size = 3, shuffle=True)

for idx, (x, y, z) in enumerate(d1):
    plt.imshow(x[1].numpy())
    plt.show()
    plt.imshow(y[1].numpy())
    plt.show()
    plt.imshow(z[1].numpy())
    plt.show()
    print(y[1])
    break
```
